```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Visualizations:

```python
# Plot the histogram
tsla_df['Close'].plot.hist(bins = 20, alpha = 0.7)

# Set the x-label, y-label, and title
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Price Frequency: Price vs Frequency')

# Show the plot
plt.show()
```

```python
# density plot with shaded area with kdeplot 'shade' parameter
sns.kdeplot(tsla_df['Close'], fill=True)
plt.show()
```

```python
# clear plots 
plt.clf()

# close line plot 
tsla_df['Close'].plot(title = 'Tesla Closing Price | 2023', figsize = (14,7))
```

```python
tsla_df['Close'].plot(kind = 'box')
```

# Close vs High vs Low

```python
tsla_df[['Close', 'High', 'Low']].plot(title = 'Tesla Data Price | 2023', figsize = (14,7))
```
