```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Diwali Sales Data.csv", encoding = 'unicode_escape')
```

```python
delete_columns = ['Status', 'unnamed1']
df = df.drop(delete_columns, axis = 1)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.isna().sum()
```

```python
# Change data type

df['Amount'] = df['Amount'].astype('int')
df['Age'] = df['Age'].astype('int')
```

```python
df.rename(columns = {'Age Group' : 'Age_Group'})
```

# Exploratory Data Analysis

```python
# Obtener el recuento de cada género
gender_counts = df['Gender'].value_counts()

# Crear el histograma con los recuentos
plt.bar(gender_counts.index, gender_counts.values, color=['pink', 'blue'])

# Configurar etiquetas y título
plt.title('Amount Female vs Male')
plt.ylabel('Amount')
plt.xlabel('Gender')

# Mostrar el histograma
plt.show()
```
